```python
#Introducir la intensidad y ocupacion de la Calle Toledo dentro del csv unido1
import pandas as pd
from os import listdir
from os.path import isfile, join
# calle elegida
df_total = pd.read_csv("calleToledo1_2021.csv",sep=";")
df_total = df_total[['id', 'fecha', 'intensidad', 'ocupacion', 'carga']]
df_total = df_total[["fecha","intensidad","ocupacion"]]
df_total

elegida = "unido1.csv"
path = "C:/Users/jonay/Desktop/Master_IoT/TFM/Callescercanas/Merge"
df_elegido2 = pd.read_csv(join(path, elegida),parse_dates=True)
df_elegido2

df_total = df_total.merge(df_elegido2,how="inner",on=["fecha"])
     
df_total
```

```python
df_total.to_csv(join(path,"Intensidad_Ocupacion_unido1.csv"),index=False)
```

## Intensidad

```python
cols = [c for c in df_total.columns if "intensidad" in c]
correlaciones = df_total[cols].corr()
correlaciones.style.background_gradient()
```

```python
correlaciones.mean()
```

```python
df_inten = df_total[cols]
df_inten.columns
```

```python
df_inten = df_inten[ ['intensidad', 'intensidad_carreraSF_2021.csv',
       'intensidad_cColegiata_2021.csv', 'intensidad_cCondeR_2021.csv',
       'intensidad_cDoctorC_2021.csv', 'intensidad_cSegovia1_2021.csv',
       'intensidad_cSegovia2_2021.csv', 'intensidad_cTintoreros_2021.csv',
       'intensidad_cToledo3_2021.csv', 'intensidad_cToledo4_2021.csv',
       'intensidad_plazaCebada_2021.csv'] ]
```

```python
from matplotlib import pyplot as plt
df_inten.columns = ["label","carreraSF","colegiata","condeR","doctorC","segovia1","segovia2","tintoreros","toledo3","toledo4","cebada"]
fig, ax = plt.subplots(figsize=(10, 6))
df_inten.loc[:100,:].plot(ax=ax)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import BayesianRidge
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import GradientBoostingRegressor

y_col = "label"
X_cols = [c for c in df_inten.columns if c!="label" ]
#Elegir para la variable y los valores de la columna atasco
y = df_inten[y_col]
#Para las X seleccionar todas las columnas que no sean ni atasco ni fecha
X = df_inten[X_cols]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.4)

metodo = RandomForestRegressor(n_estimators=200)
metodo.fit(X_train,y_train)
y_pred = metodo.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
MAE = mean_absolute_error(y_test, y_pred)
print(-MAE)
```

```python
from sklearn.metrics import mean_squared_error
rmse = mean_squared_error(y_test, y_pred, squared=False)
print(rmse)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
N= 50
ax.scatter(x=range(N),y=y_pred[:N],label="pred")
ax.scatter(x=range(N),y=y_test[:N],label="test")
for x in range(50):
    ax.plot([x,x],[y_test.values[x]-1.96*MAE,y_test.values[x]+1.96*MAE],color="green")
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(x=y_test,y=y_pred,label="pred",s=0.5)
ax.plot([0,500],[0,500],color="orange")


plt.title('Comparación entre los valores de intensidad reales y los valores predecidos')
plt.xlabel('Valor de intensidad real')
plt.ylabel('Valor de intensidad de la predicción')
```
