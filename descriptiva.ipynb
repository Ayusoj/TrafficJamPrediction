```python
import matplotlib.pyplot as plt
mediana = df3.intensidad.median()
mad = df3.intensidad.mad()
media= df3.intensidad.mean()
moda = df3.intensidad.mode()
print(mediana,mad, media, moda[0])

#df3.intensidad.hist()

plt.hist(df3.intensidad, bins=30, edgecolor='black', alpha=0.8, label='Intensidad')

# Agregar títulos y etiquetas
plt.title('Histograma de Intensidad en octubre 2021')
plt.xlabel('Intensidad')
plt.ylabel('Frecuencia')

plt.axvline(mediana, color='red', linestyle='--', label='Mediana')
plt.axvline(media, color='green', linestyle='--', label='Media')
plt.axvline(moda[0], color='pink', linestyle='--', label='Moda')

# Ajustar las divisiones en el eje x
plt.xticks(range(df3.intensidad.min(), df3.intensidad.max(), 50))


# Agregar leyenda
plt.legend()

# Mostrar el histograma
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

minimo = df3.fecha.dt.day.min()
maximo = df3.fecha.dt.day.max()
for dia in range(minimo,maximo):
    eldia = df3[df3.fecha.dt.day==dia]
    eldia.sort_values(by=["fecha"],inplace=True)
    figure, ax = plt.subplots(1,figsize=(12, 4),dpi=100) 
    outliers = eldia[(eldia.intensidad>mediana+2*mad) ]
    plt.scatter(eldia.fecha,eldia.intensidad)
    plt.scatter(outliers.fecha,outliers.intensidad,color="r")
    plt.show()
```

```python
mediana = df3.ocupacion.median()
mad = df3.ocupacion.mad()
media= df3.ocupacion.mean()
moda = df3.ocupacion.mode()
print(mediana,mad,media,moda[0])

plt.hist(df3.ocupacion, bins=30, edgecolor='black', alpha=0.7, label='Ocupación')

# Agregar títulos y etiquetas
plt.title('Histograma de Ocupación en octubre 2021')
plt.xlabel('Ocupación')
plt.ylabel('Frecuencia')

plt.axvline(mediana, color='red', linestyle='--', label='Mediana')
plt.axvline(media, color='green', linestyle='--', label='Media')
plt.axvline(moda[0], color='pink', linestyle='--', label='Moda')

# Ajustar las divisiones en el eje x
plt.xticks(range(0, 100, 10))

# Agregar leyenda
plt.legend()

# Mostrar el histograma
plt.show()
```
