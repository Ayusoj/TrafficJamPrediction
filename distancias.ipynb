Abrimos el fichero excel con las estaciones

```python
fichero = "pmed_ubicacion_12-2021.xlsx"
import pandas as pd

df = pd.read_excel(fichero)
df
```

```python
toledoSN = df[df.id == 4708]
toledoNS = df[df.id == 4709]

coord_toledoSN = (toledoSN.iloc[0].latitud, toledoSN.iloc[0].longitud)
coord_toledoNS = (toledoNS.iloc[0].latitud, toledoNS.iloc[0].longitud)
```

```python
distSN = []
distNS = []
for r in df.iterrows():
    coord = (r[1].latitud, r[1].longitud)
    distSN.append(geopy.distance.geodesic(coord_toledoSN, coord).km)
    distNS.append(geopy.distance.geodesic(coord_toledoNS, coord).km)
    
df["distSN"] = distSN
df["distNS"] = distNS

df
```

```python
df = df.reindex(columns=['tipo_elem','id','nombre','latitud','longitud','distSN','distNS'])
df
```

```python
df.sort_values(by=["distSN"],ascending=True)
```

```python
dfCercanosSN=df[df.distSN<0.43]
dfCercanosNS=df[df.distNS<0.43]
```

```python
dfCercanosSN.sort_values(by=["distSN"],ascending=True)
```

```python
from datetime import date, timedelta

start_date = date(2021, 1, 1) 
end_date = date(2021, 12, 31)    # perhaps date.now()

delta = end_date - start_date   # returns timedelta

for i in range(delta.days + 1):
    day = start_date + timedelta(days=i)
    print(day)
```
